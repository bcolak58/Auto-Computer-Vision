```python
import cv2
import matplotlib.pyplot as plt
import cvlib
from cvlib.object_detection import draw_bbox
```

```python
img=cv2.imread('cars.jpg')
```

```python
plt.imshow(img)
```

```python
box,label,count=cvlib.detect_common_objects(img) # detect_common_objects dnyadan bilinen btn objeleri tanr
```

```python
output=draw_bbox(img,box,label,count) #resimdeki objeleri tan ve kareye al , etiketle , aklama yaz
```

```python
plt.figure(figsize=(10,10))
plt.imshow(output)
```
